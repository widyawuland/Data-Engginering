```python
import requests
import pandas as pd

def source_data_from_webpage(web_page_url, table_index=0):
    try:
        df_html = pd.read_html(web_page_url)
        df_html = df_html[table_index]
    except Exception as e:
        print(f"Error membaca tabel dari {web_page_url}: {e}")
        df_html = pd.DataFrame()
    return df_html

def source_data_from_kurs(base_url, max_page=81):
    all_data = []

    for page in range(1, max_page + 1):
        url = f"{base_url}&page={page}"
        print(f"Scraping data dari: {url}")

        try:
            response = requests.get(url)
            response.raise_for_status()
            tables = pd.read_html(response.text)

            if tables:
                df_page = tables[0]
                all_data.append(df_page)
        except Exception as e:
            print(f"Error membaca data kurs dari {url}: {e}")

    if all_data:
        df_kurs = pd.concat(all_data, ignore_index=True)
    else:
        df_kurs = pd.DataFrame()

    return df_kurs

def extract_data():
    ekspor_url = "https://satudata.kemendag.go.id/data-informasi/perdagangan-luar-negeri/ekspor-non-migas-komoditi"
    impor_url = "https://satudata.kemendag.go.id/data-informasi/perdagangan-luar-negeri/impor-non-migas-komoditi"
    kurs_base_url = "https://datacenter.ortax.org/ortax/kursbi/show/USD?rentang_tanggal=2020-01-01,2024-12-31&show=USD"

    df_ekspor = source_data_from_webpage(ekspor_url)
    df_impor = source_data_from_webpage(impor_url)
    df_kurs = source_data_from_kurs(kurs_base_url, max_page=81)

    return df_ekspor, df_impor, df_kurs

def save_to_csv(df, filename):
    if not df.empty:
        df.to_csv(filename, index=False)
        print(f"Data berhasil disimpan ke {filename}")
    else:
        print(f"Tidak ada data untuk {filename}")

if __name__ == "__main__":
    df_ekspor, df_impor, df_kurs = extract_data()

    save_to_csv(df_ekspor, "ekspor_non_migas.csv")
    save_to_csv(df_impor, "impor_non_migas.csv")
    save_to_csv(df_kurs, "kurs_usd.csv")
```

# Fungsi Tranformasi
Buat fungsi untuk transformasi dengan isi meliputi

```python
def transform_data(ekspor_file, impor_file, kurs_file):
    # Import data
    df_ekspor = pd.read_csv(ekspor_file)
    df_impor = pd.read_csv(impor_file)
    df_kurs = pd.read_csv(kurs_file)

    # mengilangkan missing value
    df_ekspor.dropna(axis="columns", how="all", inplace=True)
    df_ekspor.dropna(axis="index", thresh=2, inplace=True)
    df_impor.dropna(axis="columns", how="all", inplace=True)
    df_impor.dropna(axis="index", thresh=2, inplace=True)

    # Transformasi tampilan dari lebar ke panjang(dikolom tahun)
    ekspor = df_ekspor.melt(id_vars=["HS", "URAIAN"], var_name="Tahun", value_name="Nilai_Ekspor")
    impor = df_impor.melt(id_vars=["HS", "URAIAN"], var_name="Tahun", value_name="Nilai_Impor")

    # Ubah kolom Tahun jadi integer
    ekspor["Tahun"] = ekspor["Tahun"].astype(int)
    impor["Tahun"] = impor["Tahun"].astype(int)

    # gabungkan data dengan kurs
    ekspor_kurs = ekspor.merge(df_kurs[["Tahun", "Kurs Beli"]], on="Tahun", how="left")
    impor_kurs = impor.merge(df_kurs[["Tahun", "Kurs Jual"]], on="Tahun", how="left")

    # Gabungkan ekspor dan impor
    df_gabung = ekspor_kurs.merge(
        impor_kurs,
        on=["HS", "URAIAN", "Tahun"],
        how="outer"
    )

    # rename kolom kurs
    df_gabung.rename(columns={
        "Kurs Beli": "Kurs_Beli",
        "Kurs Jual": "Kurs_Jual"
    }, inplace=True)

    # mwngurutkan kolom
    kolom_urutan = ["HS", "URAIAN", "Tahun", "Nilai_Ekspor", "Kurs_Beli", "Nilai_Impor", "Kurs_Jual"]
    kolom_ada = [col for col in kolom_urutan if col in df_gabung.columns]
    df_gabung = df_gabung[kolom_ada]

    # untuk mengisi nilai kosong
    df_gabung.fillna(0, inplace=True)

    return df_gabung
```

# Running the Workflow (ekspor,impor dan kurs)

```python
# Fungsi untuk membaca data
def read_data(ekspor_file, impor_file, kurs_file):
    df_ekspor = pd.read_csv(ekspor_file)
    df_impor = pd.read_csv(impor_file)
    df_kurs = pd.read_csv(kurs_file)
    return df_ekspor, df_impor, df_kurs

# Fungsi untuk menghapus baris/kolom null
def hapus_null(df_input):
    df_input.dropna(axis="columns", how="all", inplace=True)
    df_input = df_input.dropna(axis="index", thresh=2, inplace=False)
    return df_input

# Fungsi untuk mengubah format ekspor/impor
def ubah_format(df_input, value_name):
    df_melt = df_input.melt(id_vars=['HS', 'URAIAN'], var_name='Tahun', value_name=value_name)
    df_melt['Tahun'] = df_melt['Tahun'].astype(int)
    return df_melt

# Fungsi untuk menggabungkan ekspor dengan kurs beli
def gabung_ekspor_kurs(df_ekspor_melt, df_kurs):
    df_ekspor_melt['Tahun'] = df_ekspor_melt['Tahun'].astype(int)
    df_kurs['Tahun'] = df_kurs['Tahun'].astype(int)
    df_gabung = df_ekspor_melt.merge(df_kurs[['Tahun', 'Kurs Beli']], how='left', on='Tahun')
    return df_gabung

# Fungsi untuk menggabungkan impor dengan kurs jual
def gabung_impor_kurs(df_impor_melt, df_kurs):
    df_impor_melt['Tahun'] = df_impor_melt['Tahun'].astype(int)
    df_kurs['Tahun'] = df_kurs['Tahun'].astype(int)
    df_gabung = df_impor_melt.merge(df_kurs[['Tahun', 'Kurs Jual']], how='left', on='Tahun')
    return df_gabung

# Fungsi untuk menggabungkan ekspor & impor
def gabung_akhir(df_ekspor_kurs, df_impor_kurs):
    df_gabungan = pd.merge(
        df_ekspor_kurs,
        df_impor_kurs,
        on=['HS', 'URAIAN', 'Tahun'],
        how='outer'
    )
    return df_gabungan

# Fungsi untuk rename kolom URAIAN menjadi Uraian
def rename_kolom_uraian(df_input):
    if 'URAIAN' in df_input.columns:
        df_input.rename(columns={'URAIAN': 'Uraian', 'Kurs Beli': 'Kurs_Beli', 'Kurs Jual': 'Kurs_Jual', 'nilai_ekspor': 'Nilai_Ekspor', 'nilai_impor': 'Nilai_Impor'}, inplace=True)
    return df_input
```

# Pipeline (ekspor,impor dan kurs)
Buat fungsi Fungsi Pipeline untuk menjalankan Fungsi Pembersihan dan Transformasi

```python
# Pipeline untuk baca data
def pipeline_read(ekspor_file, impor_file, kurs_file):
    df_ekspor, df_impor, df_kurs = pd.DataFrame(), pd.DataFrame(), pd.DataFrame()
    try:
        df_ekspor, df_impor, df_kurs = read_data(ekspor_file, impor_file, kurs_file)
    except Exception as e:
        import logging
        logging.info("Baca data gagal / file tidak ditemukan")
    return df_ekspor, df_impor, df_kurs

# Pipeline untuk hapus null
def pipeline_hapus(df_ekspor, df_impor):
    try:
        df_ekspor = hapus_null(df_ekspor)
        df_impor = hapus_null(df_impor)
    except Exception as e:
        import logging
        logging.info("Hapus null gagal")
    return df_ekspor, df_impor

# Pipeline ubah format ekspor & impor
def pipeline_ubah_format(df_ekspor, df_impor):
    try:
        df_ekspor = ubah_format(df_ekspor, 'nilai_ekspor')
        df_impor = ubah_format(df_impor, 'nilai_impor')
    except Exception as e:
        import logging
        logging.info("Ubah format gagal")
    return df_ekspor, df_impor

# Pipeline gabung dengan kurs
def pipeline_gabung_kurs(df_ekspor, df_impor, df_kurs):
    try:
        df_ekspor_kurs = gabung_ekspor_kurs(df_ekspor, df_kurs)
        df_impor_kurs = gabung_impor_kurs(df_impor, df_kurs)
    except Exception as e:
        import logging
        logging.info("Gabung kurs gagal")
    return df_ekspor_kurs, df_impor_kurs

# Pipeline gabung akhir
def pipeline_gabung_akhir(df_ekspor_kurs, df_impor_kurs):
    try:
        df_gabungan = gabung_akhir(df_ekspor_kurs, df_impor_kurs)
    except Exception as e:
        import logging
        logging.info("Gabung akhir gagal")
    return df_gabungan

# Pipeline rename kolom
def pipeline_rename_uraian(df_input):
    try:
        df_output = rename_kolom_uraian(df_input)
    except Exception as e:
        import logging
        logging.info("Rename kolom URAIAN gagal")
    return df_output

# Pipeline simpan
def pipeline_simpan(df_gabungan, simpan_path):
    try:
        df_output = df_gabungan[['HS', 'Uraian', 'Tahun', 'Nilai_Ekspor', 'Nilai_Impor', 'Kurs_Beli', 'Kurs_Jual']]
        df_output = df_output.fillna(0)
        df_output.to_csv(simpan_path, index=False)
    except Exception as e:
        import logging
        logging.info("Gagal menyimpan file CSV")
    return df_output
```

# Gunakan Data ekspor, impor&kurs dan Jalankan Alur pipeline(menggabungkan)

```python
x, y, z = pipeline_read(dataeksporclean,dataimporclean,datakursclean)
x_bersih, y_bersih = pipeline_hapus(x, y)
x_isi, y_isi = pipeline_ubah_format(x_bersih, y_bersih)
x_kurs, y_kurs = pipeline_gabung_kurs(x_isi, y_isi, z)
df_gabungan = pipeline_gabung_akhir(x_kurs, y_kurs)
df_gabungan = pipeline_rename_uraian(df_gabungan)
df_gabungan['HS'] = df_gabungan['HS'].astype(str) + '.' + df_gabungan['Tahun'].astype(str)
tabel_finis = pipeline_simpan(df_gabungan, '/content/data_pipeline_gabungan.csv')

print(tabel_finis.head(10))
```

# Cek kurs mana yang berpenegaruh dengan menggunkan regresi

```python
from sklearn.linear_model import LinearRegression
import pandas as pd
import numpy as np

df = pd.read_csv('/content/data_pipeline_gabungan.csv')

tahun_list = [2020, 2021, 2022, 2023, 2024]

hasil_analisis = []

for tahun in tahun_list:
    df_tahun = df[df['Tahun'] == tahun].dropna(subset=['Nilai_Ekspor', 'Nilai_Impor', 'Kurs_Beli', 'Kurs_Jual'])

    if not df_tahun.empty:
        # Ekspor
        X_ekspor = df_tahun[['Kurs_Beli', 'Kurs_Jual']]
        y_ekspor = df_tahun['Nilai_Ekspor']
        model_ekspor = LinearRegression().fit(X_ekspor, y_ekspor)
        coef_beli_ekspor = model_ekspor.coef_[0]
        coef_jual_ekspor = model_ekspor.coef_[1]

        dominan_ekspor = 'Kurs_Beli' if abs(coef_beli_ekspor) > abs(coef_jual_ekspor) else 'Kurs_Jual'

        hasil_analisis.append({
            'Tahun': tahun,
            'Jenis': 'Ekspor',
            'Regresi': dominan_ekspor,
            'Koef_Beli': coef_beli_ekspor,
            'Koef_Jual': coef_jual_ekspor
        })

        # Impor
        X_impor = df_tahun[['Kurs_Beli', 'Kurs_Jual']]
        y_impor = df_tahun['Nilai_Impor']
        model_impor = LinearRegression().fit(X_impor, y_impor)
        coef_beli_impor = model_impor.coef_[0]
        coef_jual_impor = model_impor.coef_[1]

        dominan_impor = 'Kurs_Beli' if abs(coef_beli_impor) > abs(coef_jual_impor) else 'Kurs_Jual'

        hasil_analisis.append({
            'Tahun': tahun,
            'Jenis': 'Impor',
            'Regresi': dominan_impor,
            'Koef_Beli': coef_beli_impor,
            'Koef_Jual': coef_jual_impor
        })

    else:
        hasil_analisis.append({
            'Tahun': tahun,
            'Jenis': 'Ekspor',
            'Regresi': 'Data tidak cukup',
            'Koef_Beli': np.nan,
            'Koef_Jual': np.nan
        })
        hasil_analisis.append({
            'Tahun': tahun,
            'Jenis': 'Impor',
            'Regresi': 'Data tidak cukup',
            'Koef_Beli': np.nan,
            'Koef_Jual': np.nan
        })

df_hasil = pd.DataFrame(hasil_analisis)
df_hasil.to_csv('/content/pengaruh_kurs.csv', index=False)

print(df_hasil)
```
